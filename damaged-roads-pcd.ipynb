```python
import cv2 as cv
leaves2 = cv.imread('../input/jalananjk/jalan.jpg',0)
leaves3 = cv.imread('../input/jalananjk/jalanan.jpg',0)
png = cv.Canny(leaves2,200,250)
png2 = cv.Canny(leaves3,50,100)
```

```python
plt.subplot(121),plt.imshow(leaves2,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(png,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

Mendeteksi jalan yang rusak dengan Edge Image

**Mengurangi kabut pada gambar**
> Menggunakan Powerlaw untuk menerangkan beserta sharp filter agar kabut semakin berkurang dan gambar tetap cerah

```python
#image declaration
img = cv.imread('../input/jalananjk/jalanan.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)


#powerlaw to dehaze proccess
for powerlaw in[1.8]:
    out = np.array(255*(img/255)**powerlaw, dtype='uint8')


#2D matrix filter for better dehazing
dehaze_filter = np.array([[1, -2, -2.5],
                    [-1,  9.3, -1],
                    [-1, 1, -1]])
#2D filter for better dehazing after power law
dehaze = cv.filter2D(out, -1, dehaze_filter)


#image show
plt.figure(figsize=(7,7))
plt.imshow(img)
plt.title('Original')
plt.figure(figsize=(7,7))
plt.imshow(dehaze)
plt.title('Dehazed Image')
plt.figure(figsize=(7,7))
```

**Mendeteksi Lubang dengan Canny dan Sobel**
> Canny dan sobel akan mendeteksi tepian yang ada pada gambar setelah kabut berkurang dan pada gambar canny akan **terlihat garis** yang itu merupakan lubang, sedangkan pada gambar sobel akan terlihat **warna putih** pada gambar yang itu merupakan lubang.

```python
plt.subplot(121),plt.imshow(leaves3,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(png2,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#edge detection
lap = cv.Sobel(dehaze,cv.CV_64F,0,1, ksize=3)
lap = np.uint8(np.absolute(lap))
can = cv.Canny(dehaze, 50,255)
```

```python
plt.imshow(dehaze)
plt.title('Dehazed Image')
plt.figure(figsize=(7,7))
plt.imshow(can)
plt.title('Pothole Detection')
plt.figure(figsize=(7,7))
plt.imshow(lap)
plt.title('Pothole Detection')
```
